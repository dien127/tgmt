```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
# ng dn nh
my_path = 'D:/img/mot.jpg'
# c nh
im = Image.open(my_path)
plt.imshow(im)
```

```python
# Chuyn sang nh xm
im_gray = im.convert('L')
plt.imshow(im_gray, cmap = 'gray')
```

```python
im_array = np.array(im_gray)
print(im_array)
```

```python
# To 1 hnh v mi
plt.figure()
plt.hist(im_array.flatten(), bins=128)
plt.title('Biu  Histogram ca nh')
plt.xlabel('Gi tr ca pixel')
plt.ylabel('S lng')
plt.show()
```

```python
# Lm ni bt ng bin
plt.figure()
plt.gray()
plt.contour(im_gray, origin="image")
plt.show()
```

```python
# ng dn nh
my_path = 'D:/img/mot.jpg'

# c nh v hin th nh vi trc ta 
im = Image.open(my_path).convert("L")

# Hin th nh trn trc ta 
plt.imshow(im)

# To mt hnh v mi
plt.figure()

# t mu sc l xm
plt.gray()

# V biu  ng bin ( contour) ca nh
plt.contour(im, origin = "image")

# m bo t l trc x v trc y trn biu  l bng nhau
plt.axis('equal')

# Hin th nh
plt.show()
```
