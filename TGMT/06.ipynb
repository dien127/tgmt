```python
#import thư viện
from PIL import Image
from imgtools import *
```

```python
#Đường dẫn ảnh
my_path = 'D:/img/mot.jpg'
```

```python
# Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
# Chuyển từ ảnh RGB sang ảnh xám
gray_image = img.convert("L")
display(gray_image)
```

```python
# Phân RGB thành 3 tầng màu khác nhau
red_band, black_band, blue_band = img.split()
display(red_band)
display(black_band)
display(blue_band)
```

```python
# Trộn ảnh
merged_image =  Image.merge("RGB", (red_band, black_band, blue_band))
display(merged_image)
```
