```python
#import thư viện
from PIL import Image
from imgtools import *
```

```python
#Đường dẫn ảnh
my_path = 'D:/img/mot.jpg'
```

```python
# Đọc ảnh
img = load_image(my_path)
```

```python
display(img)
```

```python
# Định vị khu vực cắt
region = (100, 75, 175, 175)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán ảnh
# Xác dịnh vị trí cần dán
paste_position = (250, 250)
img.paste(cropped_image, paste_position)
display(img)
```
